## 인덱싱

```python
# 리스트 자료형 생성
list_data = [10, 20, 30, 40]
print(list_data)

# 첫 번째 성분 원소 인덱싱 -> 변수[인덱스]
first_element = list_data[0]
print(f'첫 번째 성분 원소 인덱싱: {first_element}')

# 네 번째 성분 원소 인덱싱: 마지막 위치 -> 인덱스 -1
# last_element = list_data[3]
last_element = list_data[-1]
print(f'네 번째 성분 원소 인덱싱: {last_element}')
```

```python
# 인덱싱 연습

## 리스트 자료형 생성
list_data1 = [1, 2, 3, ['a', 'b', 'c']]
print(list_data1)

## 첫 번째 성분 원소 인덱싱
print(list_data1[0])

## 두 번째 성분 원소 인덱싱
print(list_data1[1])

## 세 번째 성분 원소 인덱싱
print(list_data1[2])

## 네 번째(마지막) 성분 원소 인덱싱
list_data2 = list_data1[-1]
print(list_data2)

## list_data2 -> 리스트 자료형 -> 인덱싱 가능 -> 두 번째 성분 원소 인덱싱
output = list_data2[1]
print(output)

## list_data2 = list_data1[-1]
output = list_data1[-1][1]
print(output)
```

## 사용자 정의 함수

```python
# 사용자 정의 함수로서 덧셈 함수 정의
def add(x, y):
    sum = x+y
    return sum
```

```python
# 함수 사용 -> 호출

## parameter(명시 X) + argument 입력
output1 = add(3, 5)
print(output1)

## parameter(명시 O) + argument 입력
output2 = add(x=3, y=5)
print(output2)
```

```python
# 문제 1 보충: 여러 개를 동시에 추가하는 방법(2)

## 빈 리스트 생성
cart = []

## 추가 할 물품 리스트 생성
foods = ['계란','우유','사과']

## extend 함수 -> 작은 리스트를 연결 -> 긴 리스트 1개를 생성
cart.extend(foods)
print(cart)

## 물품 목록 개수 확인
count = len(cart)
print(f'장바구니에 담긴 물품의 개수 : {count}개입니다')

# 즉, extend 함수는 확장하다의 의미로 그 형식에 추가한다.
```

```python
# 문제 1 보충: 여러 개를 동시에 추가하는 방법(3)

## 빈 리스트 생성
cart = []

## 추가 할 물품 리스트 생성
foods = ['계란','우유','사과']

## + 키(덧셈 연산자) 이용 -> 작은 리스트를 연결 -> 긴 리스트 1개를 생성 -> extend 함수와 동일한 결과
cart1 = cart + foods
print(cart1)

## 물품 목록 개수 확인
count = len(cart1)
print(f'장바구니에 담긴 물품의 개수 : {count}개입니다')

## 즉 +함수는 리스트와 리스트를 더하기 때문에 extend 함수와 같다
```

### 문자열 전용 함수

```python
# split 함수

## 보통 변수.함수(값) <<<< 이지만 JOIN은 '기준값'.join(변수)

## 기본 설정 : 빈칸(공백)을 기준으로 문자열 분리
example1 = "zero one two three"
output1 = example1.split()
print(output1)

print('-'* 80)

## 쉼표(',') 기준으로 문자열 분리
example2 = "zero,two,three"
output2 = example2.split(sep=',') # sep은 매개변수가 첫번째 자리일땐 쓰지 않아도 됨. 즉, 넣으면 첫번째 자리
print(output2)
```

```python
# join 함수

## 분리된 문자열 리스트 생성
colors = ['red','blue','green','yellow']

## 연결 -> 방법 -> 공백(1칸)을 기준으로 연결
output1 = ' '.join(colors)
print(f'공백 기준으로 연결 : {output1}')

print('-'* 80)

## 연결 -> 방법 -> 쉼표 + 공백(",")을 기준으로 연결
output2 = ", ".join(colors)
print(f'쉼표 공백을 기준으로 연결 : {output2} ')
```

```python
# 블리언 자료형의 특징

## 블리언 자료형 생성
bool_data1 = True
print(bool_data1)
print('-'* 80)
bool_data2 = False
print(bool_data2)

print('-'* 80)

## 블리언 자료형을 이용한 연산
sum = bool_data1 + bool_data2 + bool_data1 # True = 1, False = 0
print(sum)
```

## for 반복문

```python
# for문 실습(1)

## 리스트 자료형 생성
list_food = ['햄버거','치킨','피자']

## for문 정의 및 실행
for food in list_food:
  # print(food)
  print('hi')
```

```python
# for문 실습(2)

## 문자열 데이터 생성
string = "안녕하세요"

## for문 정의 및 실행
for char in string:
  # print(char)
  print("hi")
```

```python
# 보충 : range(start, stop)함수 -> start부터 stop-1까지 범위에 해당하는 1씩 증가하는 정수 생성

## 0부터 99까지 1씩 증가하는 정수 생성
numbers = range(0,100)
print(numbers)

print('-'* 80)

## for문 + range 함수
for number in range(0,100):
  print(number)
```
